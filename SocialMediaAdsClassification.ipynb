```python
import numpy 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/social.csv")
print(data.head())
```

```python
print(data.describe())
print(data.isnull().sum())
```

```python
plt.figure(figsize=(15,10))
plt.title("Produto comprado por pessoas por meio do marketing em social midia")
sns.histplot(x="Age", hue="Purchased", data=data)
plt.show()
```

```python
plt.title("Produto comprado por pessoas de acordo com sua renda")
sns.histplot(x="EstimatedSalary", hue="Purchased", data=data)
plt.show()
```

```python
x = np.array(data[["Age", "EstimatedSalary"]])
y = np.array(data[["Purchased"]])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y , test_size=0.10, random_state=42)

model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
predictions = model.predict(xtest)
```

```python
print(classification_report(ytest, predictions))
```
